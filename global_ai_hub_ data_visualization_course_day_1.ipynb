```python
sns.lineplot(data=iris,y="petal_length",x=iris.index)
```

```python
sns.lineplot(data=iris,y="petal_length",x=iris.index,hue="species")
```

# Scatter Plot

```python
fig,ax=plt.subplots()
ax.scatter(iris['sepal_length'],iris['sepal_width'])
ax.set_title("İris Dataset")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
from numpy.random import randn
import matplotlib.pyplot as plt
x=20*randn(1000)+100
y=x+(10*randn(1000)+50)

plt.scatter(x,y)
plt.show()
```

# Bar Plot

```python
iris
```

```python
iris.groupby(by='species').mean()['petal_length']
```

```python
data=iris.groupby(by='species').mean()['petal_length']
data.values
```

```python
data=iris.groupby(by='species').mean()['petal_length']
ax=sns.barplot(x=data.index,y=data.values)

ax.set_title('Petal Length Means')
ax.set_xlabel("Species")
ax.set_ylabel("Mean")
plt.show()
```

```python
#Penguins Dataset
import seaborn as sns
import matplotlib.pyplot as plt

penguins=sns.load_dataset("penguins")
penguins.head()
```
