# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])
labels=["Apples","Bananas","Cherries","Dates"]

plt.pie(y,labels=labels,shadow=True)
plt.show()
```

# Heatmap

```python
from sklearn import datasets
import pandas as pd
import seaborn as sns

iris=sns.load_dataset('iris')
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.heatmap(iris.corr(),annot=True)
```

# Box Plot

```python
plt.subplot(2,1,1)
sns.boxplot(y="sepal_length",x='species',data=iris)
plt.subplot(2,1,2)
sns.boxplot(y="petal_length",x='species',data=iris)
```

```python
plt.subplot(2,1,1)
sns.boxplot(y="petal_width",x='species',data=iris)
plt.subplot(2,1,2)
sns.boxplot(y="sepal_width",x='species',data=iris)
```
