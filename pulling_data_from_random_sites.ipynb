```python
import requests
from bs4 import BeautifulSoup

headers_param={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.80 Safari/537.36 Edg/98.0.1108.50"}
glassdor=requests.get("https://www.glassdoor.com/List/Best-Jobs-in-America-2020-LST_KQ0,25.htm",headers=headers_param)

print(glassdor.status_code)
```

```python
print(glassdor.content)
```

```python
jobs=glassdor.content

soup=BeautifulSoup(jobs,"html.parser")

print(soup.title)
```

```python
print(soup.find("h1"))
```

```python
print(soup.find("h1").text)
```

```python
print(soup.find_all("div"))
```

```python
print(soup.find_all("a"))#Finding all <a> labels in site
```

```python
print(soup.find_all("p",{"class":"h2 m-0 entryWinner pb-std pb-md-0"}))#Finding all <p> labels which have "h2 m-0 entryWinner pb-std pb-md-0" class name in site 
```

```python
all_jobs=soup.find_all("p",{"class":"h2 m-0 entryWinner pb-std pb-md-0"})
for job in all_jobs:
    print(job.a.text)
```

```python
all_data=soup.find_all("div",{"class":"col-6 col-lg-4 dataPoint"})
for data in all_data:
    print(data.text)
```
