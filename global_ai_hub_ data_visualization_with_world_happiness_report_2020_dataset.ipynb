```python
cols=data.select_dtypes(exclude="object").columns#take out the type what name is 'object'
cols
```

```python
x_axis=0
y_axis=0
```

```python
from IPython.core.pylabtools import figsize
f,axes=plt.subplots(5,2,figsize=(15,20))#Making subplot name of f and axes, 5 Rows and 2 columns
f.tight_layout(pad=8)#Space between two subplots it have 8 pad
f.suptitle('Distrubution before preprocessing')#Upper Title

x_axis=0
y_axis=0

for col in cols:
    sns.histplot(data=data,x=col,kde=True,ax=axes[x_axis,y_axis])
    axes[x_axis,y_axis].set_xlabel(col.title())
    axes[x_axis,y_axis].set_ylabel(col.title())
    axes[x_axis,y_axis].set_title(f"{col.title()} Count")

    if y_axis==1:
        y_axis=0
        x_axis+=1
    else:
        y_axis+=1
        
plt.savefig("Plots/Histogram.png")
plt.show()
```

```python
f,axes=plt.subplots(5,2,figsize=(15,20))#Making subplot name of f and axes, 5 Rows and 2 columns
f.tight_layout(pad=8)#Space between two subplots it have 8 pad
f.suptitle('Distrubution before preprocessing')#Upper Title

x_axis=0
y_axis=0

for col in cols:
    sns.boxplot(data=data,x=col,ax=axes[x_axis,y_axis])
    axes[x_axis,y_axis].set_xlabel(col)
    axes[x_axis,y_axis].set_ylabel("Count")
    axes[x_axis,y_axis].set_title(f"{col.title()} Box Count")

    if y_axis==1:
        y_axis=0
        x_axis+=1
    else:
        y_axis+=1
        
plt.savefig("Plots/Boxplot.png")
plt.show()
```

```python
year_group=data.groupby("year").sum()
year_group
```

```python
year_group["Positive affect"].plot()
plt.savefig("Plots/PositiveAffectPlot.png")
```

```python
year_group["Negative affect"].plot()
plt.savefig("Plots/NegativeAffectPlot.png")
```

```python
ax1=sns.barplot(x=year_group.index,y=year_group['Social support'].values)
ax1.tick_params(axis="x",rotation=90)
plt.savefig("Plots/SocialSupportBarplot.png")
```

```python
sns.set(rc={'figure.figsize':(15,10)})
plt.title("Correlation Matrix")
sns.heatmap(data.corr(),annot=True)
plt.savefig("Plots/CorrelationMatrixHeatmap.png")
```

```python
sns.jointplot(data=data,x="year",y="Social support")
sns.jointplot(data=data,x="Life Ladder",y="Social support")
```
